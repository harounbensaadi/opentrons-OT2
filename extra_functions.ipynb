### Generating custom volume distribution

```python
import itertools
import numpy as np
import random
import math

def sumVolumes(data_list):
    volumes_sum = []
    for item in data_list:
        volume_sum = sum(item.values())
        volumes_sum.append(volume_sum)
    return volumes_sum

def createWellPlateDictionary(data, well_distribution_cells):    

    keys = []
    result = {}
    
    for key in well_distribution_cells.keys():
        keys.extend(well_distribution_cells[key])
        
    for i in range(len(keys)):
        result[keys[i]] = data[i]
    return result


def generatePercentageCombinations(needed_combinations, depth = 2):
    num_combinations = 10000 # MAX limit

    ori_ratios = {"PEG": 130/200*100, "LiAc": 20/200*100, "ssDNA": 30/200*100, "plasmid": 20/200*100}
    liquid_ranges = {"PEG": [], "LiAc": [], "ssDNA": [], "plasmid": []}

    # PEG
    step = 15
    liquid_ranges["PEG"] = np.arange(ori_ratios["PEG"]-depth*step, ori_ratios["PEG"]+depth*step+1, step).tolist()
    liquid_ranges["PEG"].pop()
    
    # LiAc
    step = 5
    liquid_ranges["LiAc"] = np.arange(ori_ratios["LiAc"]-depth*step, ori_ratios["LiAc"]+depth*step+1, step).tolist()

    # ssDNA
    step = 5
    liquid_ranges["ssDNA"] = np.arange(ori_ratios["ssDNA"]-depth*step, ori_ratios["ssDNA"]+depth*step+1, step).tolist()
    #liquid_ranges["ssDNA"].pop()

    # plasmid_DNA
    step = 3
    liquid_ranges["plasmid"] = np.arange(ori_ratios["plasmid"]-depth*step, ori_ratios["plasmid"]+depth*step+1, step).tolist()
    #liquid_ranges["plasmid"].pop()

    print("The original percentage for the liquids is:\n")
    for key, value in ori_ratios.items():
        print(f"{key}: {value} %")

    print("\nThe generated percentage ranges for the liquids are:\n")
    for key, value in liquid_ranges.items():
        print(f"{key}: {value} %")
    
    combinations = []
    liquids = list(liquid_ranges.keys())
    percentages = [liquid_ranges[liquid] for liquid in liquids]
    
    tolerance = 0.05
    for combination in itertools.product(*percentages):
        if np.isclose(np.sum(combination), 100.0, rtol=tolerance, atol=tolerance):
            
            rounded_combination = np.round(combination, decimals=1)
            if np.all(np.array(rounded_combination) <= 100):
                combinations.append(dict(zip(liquids, rounded_combination)))
             
        if len(combinations) == num_combinations:
            break
    
    print(f"\nThe max number of combinations with {round(tolerance*100, 2)} % tolerance is: {len(combinations)}")

#     # Print the combinations
#     for i, combination in enumerate(combinations, start=1):
#         print(f"Combination {i}: {combination}")

    # Sample randomly and uniformely the combinations based on input variable needed_combinations
    if needed_combinations > len(combinations):
        needed_combinations = len(combinations)
        
    sampled_combinations = random.sample(combinations, needed_combinations)
    
    # Sorting in ascending order 
    sorted_sampled_combinations = sorted(sampled_combinations, key=lambda x: (x['PEG'], x['LiAc'], x['ssDNA'], x['plasmid']))
    
    # Converting percentages to volumes in µL
    volumes_combination = []
    for item in sorted_sampled_combinations:
        modified_item = {}
        for key, value in item.items():
            if isinstance(value, (int, float)):
                modified_item[key] = math.ceil(value/100 * 200) # 200 µL 
            else:
                modified_item[key] = value
        volumes_combination.append(modified_item)

    # Modifying the volumes so that the total will be always less than 200
    # the extra is remomved from the max of the liquid inside the combination
    
    for values in volumes_combination:
        total_volume = sum(values.values())
        if total_volume > 200:
            max_volume = max(values.values())
            extra_amount = total_volume - 200
            for liquid, volume in values.items():
                if volume == max_volume:
                    values[liquid] -= extra_amount
                    break
    
#     print(f"\nThe {needed_combinations} sampled volume combinations are:\n")
        
    return volumes_combination

def generate_combined_json(output, input_json):
    combined_json = {}
    for key in output:
        combined_json[key] = {**input_json, **output[key]}
    return combined_json


def generateCustomVolumeDistribution(well_distribution_cells, input_json, depth = 2):
    volumes_combination_ = generatePercentageCombinations(len(well_distribution_cells)+1, depth)
    custom_volume_distribution = createWellPlateDictionary(volumes_combination_, well_distribution_cells)
    
    combined_custom_volume_distribution = generate_combined_json(custom_volume_distribution, input_json)
    print("\ncustom_volume_distribution = {")    

    for i, (key, value) in enumerate(combined_custom_volume_distribution.items()):
        print(f'"{key}": {value}', end='')
        if i < len(combined_custom_volume_distribution) - 1:
            print(',')
            pass
    print("\n}")

```
